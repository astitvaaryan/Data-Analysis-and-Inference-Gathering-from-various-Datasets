```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.stats import linregress
```

**Question 1: Can we classify players based on their match outcomes (wins/losses) using a logistic regression model incorporating their performance statistics in the tournament? (FrenchOpen-men-2013)**

```python

X = fm[['FSP.1', 'FSW.1', 'SSP.1', 'SSW.1', 'ACE.1', 'DBF.1', 'WNR.1', 'UFE.1', 'BPC.1', 'BPW.1', 'NPA.1', 'NPW.1',
         'FSP.2', 'FSW.2', 'SSP.2', 'SSW.2', 'ACE.2', 'DBF.2', 'WNR.2', 'UFE.2', 'BPC.2', 'BPW.2', 'NPA.2', 'NPW.2']]

y = fm['Result']

X = X.dropna()
y = y[X.index]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print('Accuracy:', accuracy)

labels = ['Player 2 Wins', 'Player 1 Wins']
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,5))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels)
plt.yticks(tick_marks, labels)
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, '{}\n{}'.format(cm[i, j], ['True Negative', 'False Positive', 'False Negative', 'True Positive'][i * 2 + j]), 
             ha="center", va="center", color="white" if cm[i, j] > thresh else "black")
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()

```

**Question 4: Can we identify any distinct groups of players based on their serving statistics, 
such as first serve percentage and number of aces, using clustering with PCA? (USOpen-women-2013)**

```python

X = uw[['FSP.1', 'ACE.1', 'FSW.1', 'SSP.1', 'SSW.1', 'DBF.1']].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=3, n_init='auto', random_state=42)
kmeans.fit(X_pca)

sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=kmeans.labels_, palette='Set2')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Clustering based on Serving Statistics')
plt.show()
```

**Question 6: Do players who win the first set have a higher chance of winning the match? (Wimbledon-men-2013)**

```python
  
wm['FirstSetWinnerWins'] = (wm['FNL.1'] > wm['FNL.2']) & (wm['ST1.1'] > wm['ST1.2'])

first_set_wins = wm[wm['ST1.1'] > wm['ST1.2']]
first_set_win_percentage = len(first_set_wins[first_set_wins['FirstSetWinnerWins'] == True]) / len(first_set_wins) * 100

print('Percentage of matches won by players who won the first set:', first_set_win_percentage, '%')

import matplotlib.pyplot as plt

labels = ['First Set Winner', 'Other']
sizes = [first_set_win_percentage, 100 - first_set_win_percentage]
colors = ['#1f77b4', '#ff7f0e']
explode = (0.1, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.title('Percentage of Matches Won by Players Who Won the First Set')
plt.show()
```

**Question 7: Can we predict a player's second serve winning percentage based on their first serve winning percentage, 
and if so, how strong is the relationship between the two variables? (AusOpen-women-2013)**

```python

corr_coef = aw['FSW.1'].corr(aw['SSW.1'])

plt.scatter(aw['FSW.1'], aw['SSW.1'])
plt.xlabel('First Serve Won Percentage')
plt.ylabel('Second Serve Won Percentage')
plt.title('Correlation between First Serve Won and Second Serve Won')

slope, intercept, r_value, p_value, std_err = linregress(aw['FSW.1'], aw['SSW.1'])
plt.plot(ww['FSW.1'], intercept + slope * ww['FSW.1'], 'r')
ticks = range(0, 51, 10)
plt.xticks(ticks)
plt.yticks(ticks)
plt.legend(['y={:.2f}x+{:.2f}'.format(slope, intercept), 'Data points'])
plt.show()

print('Correlation Coefficient: {:.2f}'.format(corr_coef))
print('Slope: {:.2f}'.format(slope))
print('Intercept: {:.2f}'.format(intercept))
print('R-squared: {:.2f}'.format(r_value**2))
```
